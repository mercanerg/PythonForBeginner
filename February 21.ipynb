```python
usa = 'WASHINGTON'
newzelland = 'WELLINGTON'
usa_set = set(usa)
newzelland_set = set(newzelland)
print('set usa', usa_set)
print('set newzelland', newzelland_set)
print('intersection ',usa_set.intersection(newzelland_set))
print('union', usa_set.union(newzelland_set))
print('difference', usa_set.difference(newzelland_set))
```
