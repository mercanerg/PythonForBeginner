```python
text = 'Microsoft workshop: Cloud Solutions + Data & AI'\
        'Microsoft Azure Academy + work placement 28 February 2022'
vowels_count = 0
constants_count = 0
numbers_count = 0
space_count = 0
symbols_count = 0
for n in text:
  if n.isalpha():
    if n in ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']:
      vowels_count += 1
    else:
      constants_count += 1
  elif n.isalnum():
    numbers_count += 1
  elif n == ' ':
    space_count += 1
  else:
    symbols_count += 1

result = f"in the text, {vowels_count} vowels {constants_count} constants\n"\
         f"{numbers_count} digits {space_count} spaces {symbols_count} "\
         f"symbols"
print(result)
      
```
